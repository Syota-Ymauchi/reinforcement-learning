### 3.2.2 ニューロンモデル

```python
import numpy as np
import matplotlib.pyplot as plt
import japanize_matplotlib
```

```python
plt.rcParams['figure.figsize'] = (7, 4)
plt.rcParams['axes.grid'] = True
```

```python
# データの作成
n = 30 # データ数
noice = 5 # ノイズ
xs = np.random.rand(n) * 30 # (0~1の値) * 30 の一様分布の値30個の要素
ts = (0.1 * xs ) ** 4 + xs + 50 + np.random.randn(n) * noice
x_range = (0, 30)
t_range = (30, 140)
```

```python
# シグモイド関数
def sigmoid(x):
    y = 1 / (1 + np.exp(-x))
    return y

xx = np.linspace(-8, 8, 100)
yy = sigmoid(xx)
plt.plot(xx, yy)
plt.show()
```

```python
# 特徴量1つ 入力層のみの出力
w0 = 2
w1 = -4

x = np.linspace(-8, 8, 10)
u = w0 * x + w1
y = sigmoid(u)

fig, ax = plt.subplots(1, 2)
ax[0].plot(x, u)
ax[0].set(
    title = 'linear',
    xlabel = 'x',
    ylabel = 'u'
)
ax[1].plot(x, y)
ax[1].set(
    title = 'sigmoid',
    xlabel = 'x',
    ylabel = 'y'
)
```

##### ↑このニューロンモデルを組み合わせることで出力の範囲が自由になる

### 3.2.3 ニューラルネットワークモデル

```python
# 2層ニューラルネットワーク
# ニューロンモデルを組み合わせ、出力層を線形にすることで値が自由になることを確認

w00 = 1
w01 = 3
w10 = 0.5
w11 = -1
v0 = 2
v1 = -1
v2 = 1
x = np.linspace(-8, 8, 100)
z0 = sigmoid(w00 * x + w01) # 1層目の出力
z1 = sigmoid(w10 * x + w11) # 1層目の出力
y = v0 * z0 + v1 * z1 + v2  # 出力

plt.plot(x, z0, label='1層目の出力-1')
plt.plot(x, z1, label='1層目の出力-2')
plt.plot(x, y, label='出力')
plt.legend()
plt.show()
```
